# Tratamento de valores inconsistentes/faltantes

```python
df = df[df['person_age'] > 0] # retira quem tem idade abaixo de 0 
df = df[df['person_income'] > 0] # retira quem tem income negativo
```

```python
print(df['person_age'].mean()) # imprime a media de idade
print(df['person_income'].mean()) # imprime a media de income
```

```python
df.isnull() # verifica se falta algo ( nao falta nada )
```

```python
## ver qual tem valor ausente:

df.isnull().sum()
```

## Preenchendo valores faltantes

```python

df["loan_int_rate"] = df["loan_int_rate"].fillna(df["loan_int_rate"].mean(), inplace = True)
df["person_emp_length"] = df["person_emp_length"].fillna(df["loan_int_rate"].mean(), inplace = True)
```

```python
df.loc[pd.isnull(df["loan_int_rate"])] ## nao há nada vazio mais lesgoo
```
